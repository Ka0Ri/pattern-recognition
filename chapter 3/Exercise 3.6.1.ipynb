```python
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.manifold import SpectralEmbedding
```

```python
r = 0.5
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for z in np.linspace(-1,1, 11):
    data = []
    c1, c2 = np.zeros(2)
    x1 = np.linspace((c1 - r)/4, c1 + r, 11)
    y1 = c2 + np.sqrt(r**2 - (x1 - c1)**2)
    x2 = np.linspace(c1 - r, (c1 - r)/4, 11)
    y2 = c2 - np.sqrt(r**2 - (x2 - c1)**2)
    X = np.concatenate([x1, x2])
    Y = np.concatenate([y1, y2])
    Z = z*np.ones(22)
    ax.scatter(X, Y, Z)
    
plt.show()   
        
        
```

```python
r = 0.5
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X_train = []
for z in np.linspace(-1,1, 11):
    data = []
    theta = np.arange(0.5, 2.05*np.pi, 0.2)
    r = 0.1*theta
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    z = z*np.ones_like(theta)
    X_train.append((np.array([x, y, z])).T)
    ax.scatter(x, y, z)
    
    
plt.show()   
        
```

```python
X = np.concatenate([X_train[0], X_train[1], X_train[2], X_train[3], X_train[4], X_train[5], X_train[6], X_train[7], X_train[8], X_train[9], X_train[10]])
embedding = SpectralEmbedding(n_components=2)
XT = embedding.fit_transform(X)
plt.plot(XT[:,0], XT[:,1])
n = X_train[0].shape[0]
for i in range(3):
    X_class = XT[i:(i + 1)*n]
    plt.plot(X_class[:,0], X_class[:,1])
plt.show()
    
```
