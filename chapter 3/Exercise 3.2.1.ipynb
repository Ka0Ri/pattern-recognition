```python
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

def random_MixtureGaussian(N, P, m, S, cl):
    d = m.shape[1]
    c = m.shape[0]
    z = np.random.choice(np.arange(0,c), N, p=P)
    X = np.ones([1,d])
    for i in range(0,c):
        n = np.sum(z == np.array([i]*N))
        x = np.random.multivariate_normal(m[i], S[i], n)
        X = np.concatenate([X, x])
    X = X[1:]
    Y = cl*np.array(N*[1])
    return X, Y
```

# Question 1

```python
m1 = np.array([[-6, 6, 6]])
m2 = np.array([[6, 6, 6]])
S = np.array([[[0.3, 1.0, 1.0], [1.0, 9.0, 1.0], [1.0, 1.0, 9.0]]])
P = np.array([1])
N = 200

X1, Y1 = random_MixtureGaussian(N, P, m1, S, 0)
X2, Y2 = random_MixtureGaussian(N, P, m2, S, 1)

X_train = np.concatenate([X1, X2])
Y_train = np.concatenate([Y1, Y2])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X1[:,0], X1[:,1], X1[:,2])
ax.scatter(X2[:,0], X2[:,1], X2[:,2])
plt.show()
```

```python
pca = PCA(n_components=3)
pca.fit(X_train)
print("pca components", pca.components_)
print("explained variance", pca.explained_variance_ratio_ )
```

```python
X112 = X1[:,[0, 1]]
X212 = X2[:,[0, 1]]
plt.plot(X112[:,0], X112[:,1], 'bo')
plt.plot(X212[:,0], X212[:,1], 'r+')
plt.title("X11-X12")
plt.show()
X113 = X1[:,[0, 2]]
X213 = X2[:,[0, 2]]
plt.plot(X113[:,0], X113[:,1], 'bo')
plt.plot(X213[:,0], X213[:,1], 'r+')
plt.title("X11-X13")
plt.show()
X123 = X1[:,[1, 2]]
X223 = X2[:,[1, 2]]
plt.plot(X123[:,0], X123[:,1], 'bo')
plt.plot(X223[:,0], X223[:,1], 'r+')
plt.title("X12-X13")
plt.show()
```

```python
XT1 = pca.transform(X1)
XT2 = pca.transform(X2)
X112 = XT1[:,[0, 1]]
X212 = XT2[:,[0, 1]]
plt.plot(X112[:,0], X112[:,1], 'bo')
plt.plot(X212[:,0], X212[:,1], 'r+')
plt.title("Y1 - Y2")
plt.show()
X113 = XT1[:,[0, 2]]
X213 = XT2[:,[0, 2]]
plt.plot(X113[:,0], X113[:,1], 'bo')
plt.plot(X213[:,0], X213[:,1], 'r+')
plt.title("Y1 - Y3")
plt.show()
X123 = XT1[:,[1, 2]]
X223 = XT2[:,[1, 2]]
plt.plot(X123[:,0], X123[:,1], 'bo')
plt.plot(X223[:,0], X223[:,1], 'r+')
plt.title("Y2 - Y3")
plt.show()
```
